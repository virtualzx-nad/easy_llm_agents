```python
from lm_agent.agent import Agent
model = 'vicuna'
agent = Agent(
    model=model,
    work_dir=model,
    essential_only=False, disable=['delegate', 'python', 'writer'],
    config={
#        'roles': ('System', 'Assistant'), 
        'command':{
            'delegate': {'save':True},
            'reader': {'summarization_model': model},
        }
    },
    log_file=f'{model}.log',
)
agent.instruct("""
I heard there's a conflict in Sudan in the past month but I don't know what happened.
Can you tell me what it is?
""")
```
