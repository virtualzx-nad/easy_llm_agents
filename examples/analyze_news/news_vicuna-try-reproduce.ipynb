```python
from lm_agent.agent import Agent
model = 'vicuna'
agent = Agent(
    model=model,
    work_dir=model,
    essential_only=False, disable=['delegate', 'python', 'writer'],
    config={
        'command':{
            'delegate': {'save':True},
            'reader': {'summarization_model': model},
        }
    },
    log_file=f'{model}.log',
)
agent.instruct("""
I heard a conflict just broke out in Sudan but I don't know what happened.
Can you tell me what happened?
""")
```

```python
from lm_agent.conversation import format_messages

print(agent.conversation._last_system_prompt + format_messages(agent.conversation.history, config=agent.model.config).replace('###', '\n###'))
```
